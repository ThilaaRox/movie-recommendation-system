importing the dependencies

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection and Pre-Processing

```python
# loading the data from the csv file to apandas dataframe
movies_data = pd.read_csv('/25k IMDb movie Dataset.csv')
```

```python
# printing the first 5 rows of the dataframe
movies_data.head()
```

```python
# number of rows and columns in the data frame
movies_data.shape
```

```python
# selecting the relevent features for recommendation
selected_features = ['Generes','Overview','Plot Kyeword','movie title','Rating']
print(selected_features)
```

```python
# replacing the null values with null string

for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
# combining all the 5 selected features

combined_features = movies_data['Generes']+''+movies_data['Overview']+''+movies_data['Plot Kyeword']+''+movies_data['movie title']+''+movies_data['Rating']
print(combined_features)
```

```python
# converting the text data to feature vectors
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

Cosine similarity

```python
# getting the similarity scores using cosine similarity
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
# getting the movie information from the user
movie_name = input(' Describe your movie : ')
```

```python
# creating a list all the movie names given in the dataset
list_of_all_titles = movies_data['movie title'].tolist()
print(list_of_all_titles)
```

```python
# finding the close match for the movie name given by the user
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# finding the index of the movie with title

#index_of_the_movie = movies_data[movies_data.title == close_match]['path'].values[0]
#print(index_of_the_movie)
filtered_data_exact = movies_data[movies_data['movie title'] == close_match]

# Get the index of the matching row(s)
matching_indices_exact = filtered_data_exact.index
matching_indices_exact[0]
```

```python
# getting a list of similar movies

similarity_score = list(enumerate(similarity[matching_indices_exact[0]]))
print(similarity_score)
```

```python
print(similarity_score[0][1])
```

```python
len(similarity_score)
```

```python
# sorting the movies based on their similarity score

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['movie title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```

```python
movie_name = input(' Describe your movie : ')

list_of_all_titles = movies_data['movie title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

filtered_data_exact = movies_data[movies_data['movie title'] == close_match]

matching_indices_exact = filtered_data_exact.index

similarity_score = list(enumerate(similarity[matching_indices_exact[0]]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['movie title'].values[0]
  if (i<6):
    print(i, '.',title_from_index)
    i+=1
```
